```python
# @title
def arr1inArr2(arr1, arr2):
  for i in range(len(arr2)):
    if equalArr(arr1, arr2[i]):
      return True
  return False

def equalArr(arr1, arr2):
  for i in range(len(arr1)):
    if arr1[i] != arr2[i]:
      return False
  return True
```

```python
# @title
#Density score
score9 = [0 for _ in range(25)]

#Halves comparsion score
score10 = [0 for _ in range(25)]

sumList1 = [0 for _ in range(25)]
sumList2 = [0 for _ in range(25)]
sumListT = [0 for _ in range(25)]

for i in range(len(arr)):
  sum1 = 0
  sum2 = 0
  for j in range(0, int(len(arr[i])/2)):
    sum1 += arr[i][j]
  for j in range(int(len(arr[i])/2), len(arr[i])):
    sum2 += arr[i][j]
  sumList1[i] = sum1
  sumList2[i] = sum2
  sumListT[i] = sum1 + sum2

for i in range(len(arr)):
  print(sumList1[i], sumList2[i], sumListT[i])

for i in range(len(arr)):
  if (sumListT[i] < 5):
    score9[i] += -5
  elif (sumListT[i] >= 5 and sumListT[i] <= 8):
    score9[i] += 7
  elif (sumListT[i] >= 9 and sumListT[i] <= 15):
    score9[i] += 15
  elif (sumListT[i] >= 16 and sumListT[i] <= 20):
    score9[i] += -20
  else:
    score9[i] += -40

  if (sumListT[i] % 2 == 0):
    score9[i] += 1
  else:
    score9[i] += -1

  if (arr[i][0] == 1 and arr[i][24] == 1 and sumListT[i] <= 6):
    score9[i] += -4

  if (sumList1[i] > sumList2[i]):
    score10[i] += 2
  elif (sumList1[i] < sumList2[i]):
    score10[i] += 1

print("\n")
for i in range(len(arr)):
  print(score9[i], score10[i])
```

```python
# @title
scoreTotal = [[0,0] for _ in range(25)]

for i in range(25):
  scoreTotal[i][0] += score1[i] + score2[i] + score3[i] + score4[i]
  + score5[i] + score6[i] + score7[i] + score8[i] + score9[i]
  + score10[i]
  scoreTotal[i][1] = i + 1

def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    already_sorted = True
    for j in range(n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
        already_sorted = False
    if already_sorted:
      break
  return arr

# Example usage
print(scoreTotal)
print(bubble_sort(scoreTotal))
```
